# Agno Blog Generator - Example Usage

This notebook demonstrates how to use the Agno blog generator programmatically.

```python
import sys
sys.path.append('../src')

from agno_blog.application.blog_generation_service import BlogGenerationService
from agno_blog.infrastructure.db import get_workflow_db
from agno.workflow.workflow import Workflow
```

## Initialize the Service

```python
service = BlogGenerationService()
```

## Create Workflow

```python
async def blog_generation_execution(session_state, topic: str = None, **kwargs):
    """Workflow execution function."""
    return await service.generate_blog_post(session_state, topic, **kwargs)

workflow = Workflow(
    name="Blog Post Generator",
    description="Advanced blog post generator",
    db=get_workflow_db(),
    steps=blog_generation_execution,
    session_state={},
)
```

## Generate a Blog Post

```python
topic = "The Future of Renewable Energy"

response = await workflow.arun(
    topic=topic,
    use_search_cache=True,
    use_scrape_cache=True,
    use_blog_cache=True,
)

print(response.content)
```
